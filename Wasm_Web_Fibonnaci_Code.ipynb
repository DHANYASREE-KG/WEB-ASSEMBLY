```python
%%writefile fibonacci.wat
(module
  (func $fib (param $n i32) (result i32)
    (if (result i32)
      (i32.le_s (local.get $n) (i32.const 1))
      (then (local.get $n))
      (else
        (i32.add
          (call $fib (i32.sub (local.get $n) (i32.const 1)))
          (call $fib (i32.sub (local.get $n) (i32.const 2)))
        )
      )
    )
  )
  (export "fib" (func $fib))
)
```

```python
!apt-get update -qq
!apt-get install -y wabt
!wat2wasm fibonacci.wat -o fibonacci.wasm


```

```python
%%writefile run_fibonacci.js
const fs = require('fs');

(async () => {
  const bytes = fs.readFileSync('fibonacci.wasm');
  const { instance } = await WebAssembly.instantiate(new Uint8Array(bytes));

  const input = parseInt(process.argv[2]) || 30; // Use input or default 30

  const start = process.hrtime.bigint();
  const result = instance.exports.fib(input);
  const end = process.hrtime.bigint();

  console.log(`Fibonacci(${input}) = ${result}`);
  console.log(`Execution time: ${(end - start) / 1000000n} ms`);
})();




```
